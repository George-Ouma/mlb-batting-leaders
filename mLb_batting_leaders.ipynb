```python
# The packages I will need for downloading the pages and parsing the data
import requests
from bs4 import BeautifulSoup
import  pandas as pd
import re
```

```python
page = requests.get('https://www.espn.com/mlb/history/leaders/_/breakdown/season/year/2018/start/1')
soup = BeautifulSoup(page.content, 'html.parser')
#print(soup.prettify())
```

```python
colhead = soup.find('tr', class_ = 'colhead')
#print(colhead)
columns = []
for td in colhead.find_all('td'):
    columns.append(td.get_text().strip())
    
columns
```

```python
df = pd.DataFrame(columns = columns)
df
```

```python
for tr in soup.find_all('tr', re.compile('row player-10-')):
    stats = [stat.get_text() for stat in tr.find_all('td')]
    
    #print(stats, end = '\n')
    
    temp_df = pd.DataFrame(stats).transpose()
    temp_df.columns = columns
    
    df = pd.concat([df, temp_df], ignore_index = True)
    
df.head(10)
```

```python
for i in range(1, 331, 50):
    url = 'https://www.espn.com/mlb/history/leaders/_/breakdown/season/year/2018/start/{}'.format(i) 
    page = requests.get(url) 
    soup = BeautifulSoup(page.content, 'html.parser')

    for tr in soup.find_all('tr', re.compile('row player-10-')):
        stats = [stat.get_text() for stat in tr.find_all('td')]

        #print(stats, end = '\n')

        temp_df = pd.DataFrame(stats).transpose()
        temp_df.columns = columns

        df = pd.concat([df, temp_df], ignore_index = True)

df
```
